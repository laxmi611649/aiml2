```python
import pandas as pd
import matplotlib.pyplot as plt
iris = pd.read_csv('/content/Iris (2).csv',names=['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','class'])
iris.head()
iris.tail()
iris.info()
iris.describe()
```

```python
import matplotlib.pyplot as plt

x=iris['PetalLengthCm']
y=iris['PetalWidthCm']
plt.scatter(x,y)
plt.title("scatter plot Petal length vs width")
plt.xlabel("petal length")
plt.ylabel("Petal width")
plt.show()
```

```python
Species_counts = iris['class'].value_counts()
plt.bar(Species_counts.index, Species_counts.values, color='lightblue')
plt.title("number of flowers by species")
plt.xlabel("species")
plt.ylabel("count")
plt.grid(axis='y')
plt.show()
```

```python
plt.hist(iris['PetalLengthCm'], bins=10, color='lightgreen')
plt.title("Distribution of Petal Length")
plt.xlabel("petal length")
plt.ylabel("frequency")
plt.grid(True)
plt.show()
```
